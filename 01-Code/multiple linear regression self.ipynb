```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_excel('data/train.xlsx')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
def train_test():
    features=[]
    target=[]
    
    for ten_perc,twelve_perc,coll_gpa,eng,quant,logical,salary in zip(dataset['10percentage'],
                                                                     dataset['12percentage'],
                                                                     dataset['collegeGPA'],
                                                                     dataset['English'],
                                                                     dataset['Quant'],
                                                                     dataset['Logical'],
                                                                      dataset['Salary']):
        features.append([ten_perc,twelve_perc,coll_gpa,eng,quant,logical])
        target.append(salary)
    return features,target    
```

```python
trainX,trainY = train_test()
```

```python
trainX
```

```python
trainY
```

```python
regression=LinearRegression()
regression.fit(trainX,trainY)
```

```python
regression.score(trainX,trainY)
```
