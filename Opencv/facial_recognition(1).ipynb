```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import os
```

```python
#loading all directories and name of folder as label to every image inside
files = os.listdir('./Faces/')
#print files
faces_paths=[]
labels=[]

for f in files:
    ims =os.listdir('./Faces/'+f)
    for px in ims:
        faces_paths.append('./Faces/'+f+'/'+px)
        labels.append(f)
        
print(faces_paths)
print(labels)
```

```python
imgs=[]

for fx in faces_paths:
    im=cv2.imread(fx)
    im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    imgs.append(im)
    
print(len(imgs)) 
print(im.shape)
print(imgs)
```

```python
all_faces=[]
plt.figure(0)
for mm in range(13):
    im=imgs[mm][100:550,400:800,:]   #crop image[height,length,all colors]
    print(im.shape)
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    print(img.shape)
    all_faces.append(img)
    plt.imshow(img,cmap='gray')
    plt.show()
    
for mm in range(13,23):
    im=imgs[mm][150:550,400:800,:]   #crop image[height,length,all colors]
    print(im.shape)
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    print(img.shape)
    all_faces.append(img)
    plt.imshow(img,cmap='gray')
    plt.show()    
    
for mm in range(23,36):
    im=imgs[mm][100:500,400:800,:]   #crop image[height,length,all colors]
    print(im.shape)
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    print(img.shape)
    all_faces.append(img)
    plt.imshow(img,cmap='gray')
    plt.show()    
    
```

```python
print(len(all_faces))
all_faces=np.asarray(all_faces)
print(all_faces.shape)
```

```python
lab=np.asarray(labels[:])
print(lab)
print(lab.shape)

np.save('./facelabels',lab)
np.save('./facedata',all_faces)
```
