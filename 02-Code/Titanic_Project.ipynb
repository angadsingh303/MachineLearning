```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/titanic.csv')
```

```python
df.head()
```

# Cabin Manipulations

```python
column_drop = [
    'PassengerId',
    'Ticket',
    'Name'
]
df.drop(column_drop,axis=1,inplace=True)
```

```python
func1= lambda x: str(x[0]) if not pd.isnull(x) else np.NaN
df['Cabin'] = df['Cabin'].apply(func1)
```

```python
df = pd.concat([df,pd.get_dummies(df['Cabin'],prefix="Cabin")],axis=1)
df.head()
```

# Mapping

```python
Sex_map = lambda x: 1 if str(x)=='female' else 0
df['Sex']=df['Sex'].apply(Sex_map)
df.head()
```

# Embarked

```python
df = pd.concat([df,pd.get_dummies(df['Embarked'],prefix='Emb')],axis=1)
df.head()
```

```python
df['Fare'] = df['Fare'].apply(np.around)
```
