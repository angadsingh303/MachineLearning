```python
def load_data(filename):
    dataset = []
    
    with open(filename,'r') as file:
        reader = csv.reader(file)
        
        for row in reader:
            dataset.append(row)
    
    return dataset
```

```python
def str_to_float(dataset):
    
    for row in range(len(dataset)):
        for col in range(len(dataset[0])):
            dataset[row][col] = float(dataset[row][col].strip())
```
